# 굿모닝FM 테이 라디오 오늘의 키워드 

https://minicast.imbc.com/PodCast/pod.aspx?code=1000587100000100000

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://minicast.imbc.com/PodCast/pod.aspx?code=1000587100000100000"
html = urlopen(url)
soup = BeautifulSoup(html, "html.parser")

for i in soup.find_all('item'):
    title = i.find('title').get_text(strip=True)
    mp3 = i.find('guid').get_text(strip=True)
    if '양효걸' in title or '염규현' in title or '곽우신' in title:
        date = title.find(')')
        start = title.find('"')
        end = title.rfind('"')
        print(title[:date+1]+title[start:end+1])
        print(mp3)
```
